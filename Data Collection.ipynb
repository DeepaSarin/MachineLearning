<h1> Extracting tweets using twint library in Python<h1>

```python
#Bitcoin 
import pandas as pd
import twint
import nest_asyncio

def scrap(keywords,from_date,to_date,csv_name):
    nest_asyncio.apply()
    c = twint.Config()
    c.Search = "Ukraine war bitcoin " #keyword for search
    c.Popular_tweets = True
    c.Since = from_date
    c.Until=to_date
    #c.Min_likes = 30000
    #c.Min_replies = 1000
    #c.Min_retweets = 100
    c.Limit = 10000 #limit of the number of tweets which will be extracted
    c.Store_csv = True 
    c.Output = csv_name
    twint.run.Search(c)
    return 

from_date_feb='2022-02-01'
to_date_feb='2022-02-28'
from_date_march='2022-03-01'
to_date_march='2022-03-30'

scrap("Ukraine war bitcoin ",from_date_feb,to_date_feb,'bitcoin_feb.csv')#Feb
scrap("Ukraine war bitcoin ",from_date_march,to_date_march,'bitcoin_march.csv')#March
scrap("Russia Ukraine  crypto ",from_date_feb,to_date_feb,'crypto_feb.csv')
scrap("Russia Ukraine  crypto ",from_date_march,to_date_march,'crypto_march.csv')
scrap("Ukraine war blockchain ",from_date_feb,to_date_feb,'block_feb.csv')
scrap("Ukraine war blockchain ",from_date_march,to_date_march,'block_march.csv')




```
